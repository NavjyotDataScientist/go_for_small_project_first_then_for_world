pandas.DataFrame.at — simple explanation
What is .at?

.at is used to get or set one single value in a DataFrame using row label and column label.

Think: exact location, very fast.

DataFrame.at is the fastest way to access or update a single cell using labels.

```python
import pandas as pd

df = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [100, 200, 300]
})

value = df.at[1, "B"]
print(value)
```

pandas.DataFrame.iat — simple explanation
What is .iat?

.iat is used to get or set one single value using row position and column position (numbers).

Think: index numbers, not labels.

```python
import pandas as pd

df = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [100, 200, 300]
})

value = df.iat[1, 0]
print(value)
```

pandas.DataFrame.loc — simple explanation
What is .loc?

.loc is used to select rows and columns by labels.

Think: names, not numbers.

```python
import pandas as pd

df = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [100, 200, 300]
}, index=["x", "y", "z"])

df.loc["y"]
```

pandas.DataFrame.insert() — simple explanation
What does it do?

insert() adds a new column at a specific position in a DataFrame.

Think: insert a column exactly where you want.

```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2, 3],
    "C": [7, 8, 9]
})

df.insert(1, "B", [4, 5, 6])

print(df)
```

DataFrame.__iter__() — simple explanation
What does it do?

__iter__() lets you iterate over column names in a DataFrame.

In simple words: when you loop over a DataFrame, you loop over its columns.

```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2],
    "B": [3, 4]
})

for col in df:
    print(col)
```

pandas.DataFrame.items() — simple explanation
What does .items() do?

.items() lets you iterate over columns, giving you:

the column name

the column data (Series)

Think: column name + its values

```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2],
    "B": [3, 4]
})

for col_name, col_data in df.items():
    print(col_name)
    print(col_data)
```

pandas.DataFrame.xs() — simple explanation
What does .xs() do?

.xs() means cross-section.
It is used to select data from a MultiIndex.

Think: slice data at a specific level

```python
import pandas as pd

arrays = [
    ["India", "India", "USA", "USA"],
    [2023, 2024, 2023, 2024]
]

index = pd.MultiIndex.from_arrays(arrays, names=("Country", "Year"))

df = pd.DataFrame(
    {"GDP": [3.5, 3.7, 25.5, 26.1]},
    index=index
)

print(df)
```

pandas.DataFrame.where() — simple explanation
What does .where() do?

.where() keeps values where a condition is True and replaces others with something else (default: NaN).

Think: “keep the good stuff, hide the rest”

```python
import pandas as pd

df = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [100, 200, 300]
})

# Keep values greater than 15
df2 = df.where(df > 15)
print(df2)
```

```python
import pandas as pd

df = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [100, 200, 300]
})

# Replace values greater than 15 with NaN
df2 = df.mask(df > 15)
print(df2)
```

pandas.DataFrame.query() — simple explanation
What does .query() do?

.query() lets you filter a DataFrame using a string expression, like a mini SQL for Pandas.

Think: “write the condition as text instead of using brackets”

```python
import pandas as pd

df = pd.DataFrame({
    "A": [10, 20, 30, 40],
    "B": [100, 200, 300, 400]
})

# Select rows where A > 20
df2 = df.query("A > 20")
print(df2)
```
