```python
import pandas as pd

df = pd.DataFrame({
    "age": [25, 30, None],
    "salary": [50000, 60000, None],
    "is_employed": [True, False, None]
})

print(df.dtypes)
```

```python
df2 = df.convert_dtypes()
print(df2.dtypes)
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        "age": [21, 25, 30],
        "score": [85.5, 90.0, 78.2],
        "name": ["Nav", "Alex", "Sam"]
    },
    dtype="object"   # force object dtype
)

print(df.dtypes)
```

```python
df2 = df.infer_objects()
print(df2.dtypes)
```
