What is DataFrame.combine()? (Simple English)

DataFrame.combine() means:

Take two DataFrames and combine their values cell-by-cell using your own rule (function).

Unlike + or __add__, you decide HOW to combine the values.

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    "A": [1, 5, 3],
    "B": [4, 2, 6]
})

df2 = pd.DataFrame({
    "A": [2, 3, 7],
    "B": [1, 8, 4]
})

result = df1.combine(df2, np.maximum)
print(result)
```
